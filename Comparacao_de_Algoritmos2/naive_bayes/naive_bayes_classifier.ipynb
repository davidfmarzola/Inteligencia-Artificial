```python
import pickle
with open('../sample_data/balanced_labor.pkl','rb') as f:
  X_over, X_test, y_over, y_test = pickle.load(f)
```

```python
from sklearn.naive_bayes import GaussianNB 

X_over.toarray() # converting to dense matrix

naiveBayes = GaussianNB()
naiveBayes.fit(X_over, y_over)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

previsao = naiveBayes.predict(X_test)
print(accuracy_score(y_test,previsao))
print(confusion_matrix(y_test, previsao))
print(classification_report(y_test, previsao))
```
