```python
import pandas as pd
# Biblioteca de visualização de dados
import plotly.express as px
```

```python
df = pd.read_csv("trabalhoPratico.csv", sep=';')
df
```

**Gráficos de pizza**

```python
fig = px.pie(names=df['work_type'], values = df['stroke'], width=500, height=500)
fig.update_traces(title_text='Pizza', title_position='top left')
fig.show()
```

**Biblioteca Matplotlib**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Visualizing statistical relationships**

```python
import seaborn as sns
sns.pairplot(df, hue="stroke")
```

```python
sns.relplot(data=df, x="avg_glucose_level", y="age")
```

```python
sns.relplot(data=df, x="avg_glucose_level", y="age", hue="stroke")
```

```python
sns.relplot(
    data=df,
    x="avg_glucose_level", y="age", hue="stroke", style="stroke"
)
```

```python
sns.relplot(
    data=df,
    x="avg_glucose_level", y="age", hue="stroke", style="ever_married",
)
```

**Emphasizinh continuity with line plots**

```python
sns.relplot(data=df, x="age", y="bmi", kind="line")
```

**Aggregation and representing uncertainty**

```python
sns.relplot(data=df, x="avg_glucose_level", y="bmi", kind="line")
```

**Plotting subsets of data with semantic mappings**

```python
sns.relplot(
    data=df, kind="line",
    x="age", y="bmi", hue="stroke",
)
```

**Showing multiple relationships with facets**

```python
sns.relplot(
    data=df,
    x="age", y="bmi", hue="stroke", col="hypertension",
)
```
